```python
import pandas as pd
```

```python
companies_df = pd.read_csv("comp.csv", encoding='latin1')
investments_df = pd.read_csv("inv.csv", encoding='latin1')
companies_df
```

```python
investments_df
```

```python
silicon_valley_cities = [
    "San Francisco", "Palo Alto", "San Jose", "Mountain View",
    "Menlo Park", "Redwood City", "Cupertino", "Sunnyvale"
]

companies_df['city'] = companies_df['city'].str.strip()

sv_companies_df = companies_df[companies_df['city'].isin(silicon_valley_cities)]
len(sv_companies_df)
```

```python
# Clean and prepare the investments dataset
investments_df['company_name'] = investments_df['company_name'].str.strip()
investments_df['investor_name'] = investments_df['investor_name'].str.strip()
investments_df['company_city'] = investments_df['company_city'].str.strip()

# Filter investments to only include companies in our Silicon Valley list
sv_investments_df = investments_df[investments_df['company_city'].isin([
    "San Francisco", "Palo Alto", "San Jose", "Mountain View",
    "Menlo Park", "Redwood City", "Cupertino", "Sunnyvale"
])]

# Create a mapping: investor -> list of companies they invested in
from collections import defaultdict

investor_to_companies = defaultdict(set)

# Grouping investors and their portfolio companies
for _, row in sv_investments_df.iterrows():
    investor = row['investor_name']
    company = row['company_name']
    if pd.notna(investor) and pd.notna(company):
        investor_to_companies[investor].add(company)

# Build edges between companies that share an investor
edges = set()

for companies in investor_to_companies.values():
    companies = list(companies)
    for i in range(len(companies)):
        for j in range(i + 1, len(companies)):
            edges.add((companies[i], companies[j]))

# Convert edges to DataFrame for preview
edges_df = pd.DataFrame(list(edges), columns=['Company A', 'Company B'])

len(edges_df)
```

```python
import networkx as nx

# BUild the network graph
G = nx.Graph()

# Add nodes
for company in sv_companies_df['name'].dropna().unique():
    G.add_node(company)

# Add edges (based on shared investors)
G.add_edges_from(edges)

degree_centrality = nx.degree_centrality(G)

betweenness_centrality = nx.betweenness_centrality(G)

# Number of connected components
num_components = nx.number_connected_components(G)

# Largest connected component size
largest_cc = max(nx.connected_components(G), key=len)
largest_cc_size = len(largest_cc)

# Combine centrality metrics into a DataFrame
centrality_df = pd.DataFrame({
    'Company': list(degree_centrality.keys()),
    'Degree Centrality': list(degree_centrality.values()),
    'Betweenness Centrality': list(betweenness_centrality.values())
}).sort_values(by='Degree Centrality', ascending=False)


num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

{
    "Total Startups (nodes)": num_nodes,
    "Total Connections (edges)": num_edges,
    "Connected Components": num_components,
    "Largest Component Size": largest_cc_size
}
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Visualize the Largest connected component
# extract subgrap of the largest component
largest_cc_subgraph = G.subgraph(largest_cc).copy()


sub_deg_cent = nx.degree_centrality(largest_cc_subgraph)

from networkx.algorithms.community import greedy_modularity_communities

communities = list(greedy_modularity_communities(largest_cc_subgraph))
community_map = {}
for i, comm in enumerate(communities):
    for node in comm:
        community_map[node] = i

# Generate color and size mappings
node_color = [community_map[node] for node in largest_cc_subgraph.nodes()]
node_size = [5000 * sub_deg_cent[node] for node in largest_cc_subgraph.nodes()]

# Create layout and plot
plt.figure(figsize=(14, 12))
pos = nx.spring_layout(largest_cc_subgraph, seed=42)

nx.draw_networkx_nodes(largest_cc_subgraph, pos, node_size=node_size, node_color=node_color, cmap=cm.tab20, alpha=0.8)
nx.draw_networkx_edges(largest_cc_subgraph, pos, alpha=0.2, width=0.5)

plt.title("Silicon Valley Startup Network (Largest Connected Component)", fontsize=16)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#Identify and display Top central startups from the largestcComponent

# filter the centrality df to only include companies in the largest component
top_nodes_df = centrality_df[centrality_df['Company'].isin(largest_cc)]

# Sort by degree centrality and get top 10
top_companies = top_nodes_df.sort_values(by='Degree Centrality', ascending=False).head(10)
top_companies
```

```python
import random

# We will simulate a basic SIR model on the largest connected component
def simulate_diffusion(graph, start_node, steps=10, infection_prob=0.2):
    infected = set([start_node])
    newly_infected = set([start_node])
    infection_history = [len(infected)]

    for _ in range(steps):
        next_infected = set()
        for node in newly_infected:
            neighbors = list(graph.neighbors(node))
            for neighbor in neighbors:
                if neighbor not in infected and random.random() < infection_prob:
                    next_infected.add(neighbor)
        newly_infected = next_infected
        infected.update(newly_infected)
        infection_history.append(len(infected))

    return infection_history

seed_node = top_companies.iloc[0]['Company']  # e.g., Airbnb

# Run simulation
infection_counts = simulate_diffusion(largest_cc_subgraph, start_node=seed_node, steps=15, infection_prob=0.2)

# Plot the result
plt.figure(figsize=(10, 5))
plt.plot(range(len(infection_counts)), infection_counts, marker='o')
plt.title(f"Virus-like Innovation Diffusion from {seed_node}", fontsize=14)
plt.xlabel("Time Steps")
plt.ylabel("Total Infected Startups")
plt.grid(True)
plt.tight_layout()
plt.show()
```
